## Условие 3: 3 задача ##
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.figure(figsize=(6, 4))
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg')
plt.xlabel('price')
plt.ylabel('yr_built')
```

## Вывод: ## 
Год постройки оказывает минимальное воздействие на цену дома

```python
plt.figure(figsize=(6, 4))
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg')
plt.xlabel('price')
plt.ylabel('bedrooms')
```

## Вывод: ## 
Количество спален не оказывает влияние на цену дома

```python
plt.figure(figsize=(6, 4))
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg')
plt.xlabel('price')
plt.ylabel('bathrooms')
```

## Вывод: ## 
Чем больше ванных комнат, тем выше цена на дом

```python
plt.figure(figsize=(6, 4))
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
plt.xlabel('price')
plt.ylabel('sqft_living')
```

## Вывод: ## 
Чем выше жилая площадь, тем выше цена на дом

```python
plt.figure(figsize=(6, 4))

sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('waterfront')
```
